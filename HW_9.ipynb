````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
````

````python
data_train=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tugas/train.csv')
data_train
````

````python
type(data_train)
````

````python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
le=encoder.fit(data_train.species)
labels=le.transform(data_train.species)
classes=list(le.classes_)
````

````python
jumlahclass=len(classes)
jumlahclass
````

````python
data_train=data_train.drop(['id','species'],axis=1)
data_train
````

````python
type(data_train)
````

````python
data_train['labels']=labels
````

````python
data_train.iloc[[643]]
````

````python
type(data_train)
````

````python
data_split=data_train.drop('labels',axis=1)
data_split
````

````python
x_train,x_test,y_train,y_test=train_test_split(data_split,data_train['labels'],test_size=.1,shuffle=True,stratify=labels)
````

````python
x_train
````

````python
y_train
````

**bold text**#using sgd, softmax, **sparse** categorical crossentropy ((not one hot encoded)),

````python
from keras import models
from keras import layers


nn=models.Sequential()
nn.add(layers.Dense(128,input_shape=(192,)))
nn.add(layers.Dense(jumlahclass, activation='softmax'))

nn.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
nn.summary()
````

````python
from keras.utils.vis_utils import plot_model
plot_model(nn, show_shapes=True, show_layer_names=True)
````

````python
nnhist=nn.fit(x_train, y_train, epochs=950)
````

````python
nnplot=nnhist.history['accuracy']
x=[]
for i in range(950):
  x.append(i+1)

plt.plot(x, nnplot, label = "Softmax (SGD)")
plt.legend()
plt.show()
````

better than bapaknya's result (probaby because si bapak using one output layer to accomodate sigmoid fuction, and in order to use standard categorical crossentropy)

#using sgd, sigmoid, not normalized

````python
from keras import models
from keras import layers


sigm1=models.Sequential()
sigm1.add(layers.Dense(128,input_shape=(192,)))
sigm1.add(layers.Dense(jumlahclass, activation='softmax'))
sigm1.add(layers.Dense(1, activation='sigmoid'))

sigm1.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
sigm1.summary()
````

###make for the momentum and adam first

````python
from keras import models
from keras import layers

#adam
adam1=models.Sequential()
adam1.add(layers.Dense(128,input_shape=(192,)))
adam1.add(layers.Dense(jumlahclass, activation='softmax'))
adam1.add(layers.Dense(1, activation='sigmoid'))

adam1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
adam1.summary()
````

````python
histadam=adam1.fit(x_train, y_train, epochs=950)
````

````python
from keras import models
from keras import layers
import tensorflow as tf
#sgd with momentum
opt = tf.keras.optimizers.SGD(momentum=0.9)
mom1=models.Sequential()
mom1.add(layers.Dense(128,input_shape=(192,)))
mom1.add(layers.Dense(jumlahclass, activation='softmax'))
mom1.add(layers.Dense(1, activation='sigmoid'))

mom1.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
mom1.summary()
````

````python
momenthist=mom1.fit(x_train, y_train, epochs=950)
````

###next, grab the accuracy

````python
adamplot=histadam.history['accuracy']
````

````python
momentplot=momenthist.history['accuracy']
````

````python
sigplot=sighist.history['accuracy']
````
